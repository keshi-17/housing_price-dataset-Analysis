```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gdp_percapita = pd.read_csv("datasets/housing/gdp-percapita.csv")
gdp_percapita
```

```python
better_life = pd.read_csv("datasets/housing/Better-life-index.csv")
better_life
```

```python
better_life.columns
```

```python
b = better_life.loc[(better_life['Indicator'] == 'Life satisfaction') & (better_life['Inequality'] == 'Total'), 'Value']
a =  better_life.loc[(better_life['Indicator'] == 'Life satisfaction') & (better_life['Inequality'] == 'Total'), 'Country']
index = pd.DataFrame({'Country': a , 'Value': b})
index
```

```python

gdp_data = gdp_percapita.loc[gdp_percapita['GDP per capita, current prices\n (U.S. dollars per capita)'].isin(['Hungary',  'France', 'Australia', 'United States']), '2023']
index_data = index.loc[index['Country'].isin(['Hungary','France', 'Australia', 'United States']),'Value']
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

colors = ['red', 'blue', 'green', 'yellow']
scatter = ax.scatter(x=gdp_data, y=index_data, c=colors)

ax.set(title="Scatter plot of GDP per capita vs. Life satisfaction", xlabel="GDP per capita", ylabel="Life satisfaction")

ax.legend(handles=scatter.legend_elements()[0], labels=['Hungary', 'France', 'Australia', 'United States'], title="Countries")

plt.show()


```
